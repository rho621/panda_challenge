```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data
```

```python
player_total = purchase_data.loc[:,["Gender", "SN", "Age"]]
player_total = player_total.drop_duplicates()
number_of_players = player_total.count()[0]

pd.DataFrame({"Total Players" : [number_of_players]})
```

```python
unique_item = purchase_data["Item ID"].nunique()
avg_price = round(purchase_data["Price"].mean(),2)
purchase_count = purchase_data["Purchase ID"].count()
total_revenue = purchase_data["Price"].sum()

purchase_analysis = pd.DataFrame({
                                "Number of Unique Items": [unique_item],
                                "Average Price": [avg_price],
                                "Number of Purchases": [purchase_count],
                                "Total Revenue": [total_revenue]
                                })

purchase_analysis["Average Price"] = purchase_analysis["Average Price"].astype(float).map("${:,.2f}".format)
purchase_analysis["Total Revenue"] = purchase_analysis["Total Revenue"].astype(float).map("${:,.2f}".format)
purchase_analysis
```

```python
sn = purchase_data.groupby("Gender")["SN"].nunique()
male_count = sn["Male"]
female_count = sn["Female"]
other_count = sn["Other / Non-Disclosed"]

total_player = male_count + female_count + other_count

male_percent = round(male_count/total_player *100,2)
female_percent = round(female_count/total_player *100,2)
other = round(other_count/total_player *100,2)

summary_data = pd.DataFrame({
                            "":["Male", "Female", "Other / Non-Disclosed"],
                            "Total Count": [male_count, female_count, other_count],
                            "Percentage of Players": [male_percent.astype(str) + '%', female_percent.astype(str) + '%', other.astype(str) + '%']
                            })
summary_data.set_index("")
```

```python
gender_purchase_data = purchase_data.groupby(["Gender"])["Price"].count()
avg_gender_purchase = round(purchase_data.groupby(["Gender"])["Price"].mean(),2)
total_gender_purchase = purchase_data.groupby(["Gender"])["Price"].sum()
avg_purchase_per_person = round((total_gender_purchase / sn),2)

purchase_summary_data = pd.DataFrame({
                            "Purchase Count": gender_purchase_data,
                            "Average Purchase Price": avg_gender_purchase.map("${:,.2f}".format),
                            "Total Purchase Value": total_gender_purchase.map("${:,.2f}".format),
                            "Avg Total Purchase per Person": avg_purchase_per_person.map("${:,.2f}".format),
                            })


purchase_summary_data
```

```python
bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
bin_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

player_total["Age Group"] = pd.cut(player_total["Age"], bins, labels = bin_labels)
age_total = player_total["Age Group"].value_counts()
age_percent = round(age_total / total_player * 100,2)

age_summary = pd.DataFrame({"Total Count": age_total, "Percentage of Players": age_percent.astype(str) + '%'})

```

```python
purchase_data["Age Group"] = pd.cut(purchase_data["Age"], bins, labels=bin_labels)

age_group = purchase_data.groupby("Age Group")
total_count_age = age_group["SN"].nunique()

age_purchase_count = purchase_data["Age Group"].value_counts() 
avg_age_purchase = round(age_group["Price"].mean(),2)
total_age_purchase = age_group["Price"].sum()
avg_age_total_per_person = round(total_age_purchase/total_count_age,2)

age_purchase_summary = pd.DataFrame({
                            "Purchase Count": age_purchase_count,
                            "Average Purchase Price": avg_age_purchase.map("${:,.2f}".format),
                            "Total Purchase Value": total_age_purchase.map("${:,.2f}".format),
                            "Avg Total Purchase per Person": avg_age_total_per_person.map("${:,.2f}".format),
                            })

age_purchase_summary.sort_index()

```

```python
spender_stats = purchase_data.groupby("SN")

purchase_count_spender = spender_stats["Purchase ID"].count()
avg_purchase_price_spender = round(spender_stats["Price"].mean(),2)
purchase_total_spender = spender_stats["Price"].sum()

top_spenders = pd.DataFrame({
                            "Purchase Count": purchase_count_spender,
                            "Average Purchase Price": avg_purchase_price_spender.map("${:,.2f}".format),
                            "Total Purchase Value":purchase_total_spender
                            })

top_spenders.sort_values(["Total Purchase Value"], ascending=False).head()
```

```python
items = purchase_data[["Item ID", "Item Name", "Price"]]
item_stats = items.groupby(["Item ID","Item Name"])

purchase_count_item = item_stats["Price"].count()
purchase_value = (item_stats["Price"].sum()) 
item_price = purchase_value/purchase_count_item

most_popular = pd.DataFrame({
                            "Purchase Count": purchase_count_item, 
                            "Item Price": item_price.map("${:,.2f}".format),
                            "Total Purchase Value":purchase_value
                            })
most_popular.sort_values(["Total Purchase Value"], ascending=False).head()
```
